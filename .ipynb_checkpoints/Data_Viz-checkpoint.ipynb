```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Your turn

1 - Create a 2x2 figure with matplotlib<br>
2 - Use 4 different types of plots for the following dataset

```python
# read in the aq dataset
aq = pd.read_csv('aq.csv')
```

```python
# look at the head of aq
aq.head()
```

```python
# subset the dataset to only include columns "x123" and "y1", and call the new dataframe aq1
aq1 = aq[['x123', 'y1']]
```

```python
# create a 2x2 subplot, use figsize 10x10
fig , axes = plt.subplots(2, 2, figsize = (10, 10))
```

```python
# define our x and y as a pandas series
x = aq1.x123 #from the aq1 dataframe, get me the column x123' A 'series' has one column. A pandas 'dataframe' has rows AND columns.
y = aq1['y1'] #this is different from the pandas series method! The .(columnname) method calls the column from the dataframe, while the brackets slices it from it!

# now create 4 graphs, the the upper left graph a scatter plot
# the bottom left a line plot
# the top right a histogram (just show x)
# the bottom right a bar plot


#starter code:
axes[0][0].scatter(x, y)
axes[0][1].hist(x)
axes[1][0].plot(x, y)
axes[1][1].bar(x, y)
```

```python
fig
```

```python
# Add titles to each of the axes objects
axes[0][0].set_title('Scatter plot')
axes[1][0].set_title('Line plot')
axes[0][1].set_title('Histogram')
axes[1][1].set_title('Bar chart')
```

```python
#loads tips dataset
tips = sns.load_dataset("tips") #This is a premade seaborn dataset located on a repository on Github
```

```python
tips.head()
```

```python
# visualize the relationship between bill and tip

#THIS IS ONE WAY TO DO IT

fig1, axes1 = plt.subplots(figsize = (10,10))

x1 = tips.total_bill
y1 = tips.tip

axes1.scatter(x1, y1)


#THIS IS ANOTHER WAY
# sns.scatterplot(x = tips.total_bill, y = tips.tip) #This doesn't look as nice
```

```python
# visualize the average tips men vs women pay - what's the appropriate plot?
groupby_gender = tips['tip'].groupby(tips['sex'])
groupby_gender

list(groupby_gender)

tips.groupby('sex').agg(
                        {'tip': ['mean']}
                       )
```

```python
tips.groupby('sex').agg(
                        {'tip': ['mean']}
                       )
```
